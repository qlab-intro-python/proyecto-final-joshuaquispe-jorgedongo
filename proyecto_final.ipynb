```python
import os
import re
```

La presente data recopila los discursos presidenciales desde el gobierno de Alejandro Toledo (2001) hasta el segundo año de gobierno de Pedro Castillo (2022) en formato .txt

```python
ruta_carpeta = 'discursos'
archivos = os.listdir(ruta_carpeta) 
Discursos={} 

for archivo in archivos:
  ruta_archivo = os.path.join(ruta_carpeta, archivo)
  llavenueva = archivo
  llavenueva = llavenueva.replace(".txt", "")
  Discursos[llavenueva]= None 

  with open(ruta_archivo, mode="r", encoding="utf8") as f:
    Discurso=f.read()
  Discursos[llavenueva] = Discurso

  print('Se ha procesado el archivo: ', archivo)
  print('Se ha asignado la llave: ', llavenueva)

print(len(Discursos.keys())) #
print(Discursos.keys())
```

```python

#Defino una función para convertir en minúscula los strings de una lista
def lowerIt(tokenList):
  return [i.lower() for i in tokenList]

#Tokenizo mediante espacios en blanco
def tokenEspacioBlanco(doc):
  return doc.split()

#Limpiando con REGX
def removePunctuation(string):
  punctList = [",","\\.",";","-","'",'"',"!","`","'"]
  for item in punctList:
    string = re.sub(item, "", string)
  return string

#el range te devuelve valores del 0 al 5 abierto, asi para 0 en list1 sera la "á" y procedera a cambiarla en todo el doc y así para los demás, recordad que el orden se cuenta desde el cero
def removeTildes(string):
  list1 = ["á","é","í","ó","ú","Á","É","Í","Ó","Ú"]
  list2 = ["a","e","i","o","u","A","E","I","O","U"]
  for i in range(len(list1)):
    string = string.replace(list1[i], list2[i])
  return string
#Establezco una lista de palabras que están de "relleno"
stopWords=["\\n","a", "o", "de", "la", "el", "en", "y", "que", "a", "del", "los", "las", "se", "eso", "por", "un", "para", "con", "no", "una", "su", "es", "al", "como", "lo", "fue", "más", "pero", "sus", "ha", "han", "también", "sobre", "entre", "desde", "este", "esta", "ser", "son", "muy", "nos", "sin", "hasta", "puede", "vez", "me", "mi", "yo", "ellos", "ellas", "mas"]
```

```python
#Creo un Diccionario de términos a partir de la limpieza de mis textos

Diccionario={} #creo un diccionario vacío
for llaves, values in Discursos.items(): #para cada discurso y su valor asociado de mi diccionario de Discursos
  Diccionario[llaves] = None #asigna la key de entrada a mi nuevo diccionario
  texto=((lowerIt(tokenEspacioBlanco((removeTildes(removePunctuation(Discursos[llaves]))))))) #aplico limpieza
  texto1=[x for x in texto if x not in stopWords] #elimino stop words usando list comprehensions
  Diccionario[llaves]=texto1 #añado la lista como valor a la llave creada más arriba
print(len(Diccionario.keys())) 
```

```python
Diccionario.keys()
```

```python
#Correcto para todos y separa gobiernos de Garcia

Dicc={} #Diccionario vacío
for llave, value in Discursos.items(): #para cada llave y su valor
  apellido=''.join(filter(str.isalpha, llave)) #extramos las letras alfabeticas que estan en la llave y luego las juntamos

  if apellido == 'Garcia': #si la clave extraída coincide con Garcia le añadiremos el primer carácter que indica si es su primer o segundo gobierno
    apellido=llave[0]+apellido

  if apellido not in Dicc: #si el apellido no esta en el diccionario nuevo lo introducimos como la llave y le añadimos el discurso asociado
    Dicc[apellido]=value

  else: #si ya esta simplemente añadimos un espacio y el valor asociado, así vamos juntando los discursos por presidente
    Dicc[apellido]+=' ' + value
```

```python
Dicc.keys()
```

```python
#Crea una lista de terminos usado por los presidentes limpiando el texto
Diccionario0={}
for llaves, values in Dicc.items():
  Diccionario0[llaves] = None
  texto=((lowerIt(tokenEspacioBlanco((removeTildes(removePunctuation(Dicc[llaves])))))))
  texto1=[x for x in texto if x not in stopWords]
  Diccionario0[llaves]=texto1
print(Diccionario0.keys())
```

```python
#Crea un diccionario de terminos para cada presidente
DiccTerm={}
for llave, value in Diccionario0.items(): #para cada llave y valor del diccionario por presidentes
  frecuencia = {} #creo un diccionario vacío
  for palabra in value: #para cada elemento en la lista de terminos
    if palabra in frecuencia: #si ya esta le sumo uno
      frecuencia[palabra] += 1
    else: #si no esta le asigno 1
      frecuencia[palabra] = 1

  DiccTerm[llave]=frecuencia #asigno el diccionario a la llave que es el presidente

DiccTerm.keys() #vemos las llaves creadas
```

```python
import datetime
import requests
import pandas as pd
import numpy as np
```

```python
url_base = 'https://estadisticas.bcrp.gob.pe/estadisticas/series/api/'
cod_ser = 'PD04646PD'
formato = '/json'
per = '/2004-01-19/2022-12-31'

url = f'{url_base}{cod_ser}{formato}{per}'
```

```python
response = requests.get(url)
response_json = response.json()
response_json.keys()
nombre_serie = response_json['config']['series'][0]['name']
nombre_serie
```

```python
data_raw = response_json['periods']

fecha_list = []
tc_list = []

for element in data_raw:
    fecha_list.append(element['name'])
    tc_list.append(element['values'][0])
```

```python
tc_dict = {'fechas': fecha_list, 'TC_venta': tc_list}
tc_df = pd.DataFrame(tc_dict)
tc_df['TC_venta'] = tc_df['TC_venta'].replace('n.d.', np.nan)

tc_df['TC_venta'] = tc_df['TC_venta'].astype(float)
```

```python
tc_df['fechas'] = tc_df['fechas'].astype(object)
```

```python
tc_df['fechas'] = tc_df['fechas'].str.replace('.','-')
```

```python
tc_df['fechas'] = tc_df['fechas'].apply(lambda x: x[:-3]+ '-20' + x[-2:])
```

```python
meses = {'Ene': '01', 'Feb': '02', 'Mar': '03', 'Abr': '04', 'May': '05', 'Jun': '06',
         'Jul': '07', 'Ago': '08', 'Set': '09', 'Oct': '10', 'Nov': '11', 'Dic': '12'}
```

```python
tc_df['fechas'] = tc_df['fechas'].replace(meses, regex=True)
```

```python
tc_df['fechas'] = pd.to_datetime(tc_df['fechas'], dayfirst=True)
```

```python
tc_df = tc_df.dropna()
```

```python
grafico_plot = tc_df.groupby([pd.Grouper(key='fechas', freq='1Y')])[['TC_venta']].mean()
```
